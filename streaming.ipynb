```python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def store_user_message(allMessages, text):
    user_message= {"role":"user", "content":text}
    allMessages.append(user_message)
    
def store_assistant_message(allMessages, text):
    assistant_message= {"role":"assistant", "content":text}
    allMessages.append(assistant_message)  
```

```python
allMessages = []

store_user_message(allMessages,"Can you explain Crypto Currency?")

with client.messages.stream(
    max_tokens=1000,
    messages=allMessages,
    model=model,
) as stream:
    for text in stream.text_stream:
        #print(text, end="", flush=True)
        message = stream.get_final_message()
        
print (message.content[0].text)
```
