#### Assistant prefill technique works with few models like claude-opus-4-5, claude-sonnet-4-5, claude-haiku-4-5, claude-opus-4-1

````python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-5"
````

### Link: https://platform.claude.com/docs/en/build-with-claude/handling-stop-reasons#stop-reason-values

````python
def store_user_message(allMessages, text):
    user_message= {"role":"user", "content":text}
    allMessages.append(user_message)
    
def store_assistant_message(allMessages, text):
    assistant_message= {"role":"assistant", "content":text}
    allMessages.append(assistant_message)  
    
def chat(messages, stop_sequences=None):
    params = dict(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    if stop_sequences:
        params["stop_sequences"] = stop_sequences
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
allMessages = []

store_user_message(allMessages,"Generate a sample short weather output as json")
store_assistant_message(allMessages,"```json")

answer = chat(allMessages,stop_sequences=["```"])

print (answer)
````
