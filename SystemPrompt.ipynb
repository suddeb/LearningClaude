```python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def store_user_message(allMessages, text):
    user_message= {"role":"user", "content":text}
    allMessages.append(user_message)
    
def store_assistant_message(allMessages, text):
    assistant_message= {"role":"assistant", "content":text}
    allMessages.append(assistant_message)  
    
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    ) 
    return message.content[0].text
```

```python
allMessages = []

store_user_message(allMessages, "I am an odd number. Take away one letter and I become even. What number am I?")
answer = chat(allMessages);
answer
```

#### System prompts are essential for creating AI applications that behave consistently and appropriately for their intended purpose. They transform generic AI responses into specialized, role-appropriate interactions.

```python
allMessages = []

def chat(messages, system=None):
    params = {
        "model" : model,
        "max_tokens" : 1000,
        "messages" : messages
    }
    
    if system:
        params["system"] = system
    
    # dictionary unpacking in Python
    message = client.messages.create(**params) 
    return message.content[0].text

system = """
    Consider yourself a teacher. So do not provide the answer directly, rather
    guide students to solve the problem
"""

store_user_message(allMessages, "I am an odd number. Take away one letter and I become even. What number am I?")
answer = chat(allMessages, system);
answer
```
