```python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
def store_user_message(allMessages, text):
    user_message= {"role":"user", "content":text}
    allMessages.append(user_message)
    
def store_assistant_message(allMessages, text):
    assistant_message= {"role":"assistant", "content":text}
    allMessages.append(assistant_message)  
```

```python
def chat(messages):
    params = dict(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    
    message = client.messages.create(**params)
    return message
```

```python
allMessages = []

store_user_message(allMessages, "Explain artificial intelligence in one sentence")

claudeReturnedMessage = chat(allMessages);

claudeReturnedMessage
```
