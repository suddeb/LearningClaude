```python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

#### Temperature is a decimal value between 0 and 1 that directly influences these selection probabilities. It's like adjusting the "creativity dial" on Claude's responses.

#### At low temperatures (near 0), Claude becomes very deterministic - it almost always picks the highest probability token. At high temperatures (near 1), Claude distributes probability more evenly across options, leading to more varied and creative outputs.

```python
def store_user_message(allMessages, text):
    user_message= {"role":"user", "content":text}
    allMessages.append(user_message)
    
def store_assistant_message(allMessages, text):
    assistant_message= {"role":"assistant", "content":text}
    allMessages.append(assistant_message)  
    
def chat(messages, system=None, temperature=1.0):
    params = {
        "model" : model,
        "max_tokens" : 1000,
        "messages" : messages,
        "temperature" : temperature
    }
    
    if system:
        params["system"] = system
    
    # dictionary unpacking in Python
    message = client.messages.create(**params) 
    return message.content[0].text
```

```python
allMessages = []

system = """
    Consider yourself a standup comedian. 
    Please generate jokes which is non-political and very refreshing and definitely funny
"""

store_user_message(allMessages, "Provide me a one liner morning joke")
answer = chat(allMessages, system, temperature=1.0);
answer
```
